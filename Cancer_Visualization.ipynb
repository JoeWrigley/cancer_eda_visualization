```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

import seaborn as sns
```

```python
# upload the dataset 
df = pd.read_csv('Cancer_Data.csv')
# read columns titles
df.head(10)
```

```python
# Display size of the dataset (18,208 cells)
df.size
```

```python
# There are 569 rows in this dataset
len(df)
```

```python
# number of columns = 32
num_cols = 0
for col in df:
    num_cols = num_cols + 1
print(num_cols)
```

```python
# Check the data types of each data for each column
# Outcome: only one categorical (object) variable
df.dtypes
```

```python
# Check for missiong data (for loop)
# OUTCOME: No missing data (set is clean)

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# M = Malignant & B = Benign
sns.countplot(x='diagnosis', data=df)
```

```python
df.groupby('diagnosis').mean()
```

```python
# Distribution plot for 'radius_mean'
# Most of the plots in this dataset are "right-skewed"
sns.displot(data=df.radius_mean)
sns.displot(data=df.perimeter_mean)
```

```python
# Boxplots to identify potential outliers in the data
df.boxplot('radius_mean')

```

```python
df.boxplot('perimeter_mean')
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(20,20))
sns.heatmap(correlation_matrix, cbar=True, fmt='.1f', annot=True, cmap='Blues')
plt.savefig('Correlation Heat map')
```

# Inference from EDA and Visualization
No missing Values in the dataset.
All variables have continuous numerical values (except 'diagnosis')
As evidenced by the distribution plots, the mean is slightly more than the median (right skewed) for most of the features. This is visible through the distribution plots.
As evidenced by the Count Plot, Benign cases are more than Malignant.
As evidenced in Groupby analysis, the mean of most features are clearly larger for Malignant cases compared to the benign cases 
As evidenced by Box Plot analysis, most of the features have Outliers.
Correlation Matrix reveal that many of the features are highly correlated.
